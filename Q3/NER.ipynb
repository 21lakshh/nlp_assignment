```python
import nltk
nltk.download('reuters')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

from nltk.corpus import reuters
from nltk import word_tokenize, pos_tag, ne_chunk

# Get the news article
text = reuters.raw('test/14826')
print("Article Preview:\n", text[:500])
```

```python
# Function to extract specific entities
def extract_entities(text):
    entities = {'PERSON': set(), 'ORGANIZATION': set(), 'GPE': set()} # GPE is Geo-Political Entity (Location)

    # Tokenize and Tag
    sentences = nltk.sent_tokenize(text)
    for sent in sentences:
        chunks = ne_chunk(pos_tag(word_tokenize(sent)))
        for chunk in chunks:
            if hasattr(chunk, 'label'):
                label = chunk.label()
                entity_name = ' '.join(c[0] for c in chunk)
                if label in entities:
                    entities[label].add(entity_name)
    return entities

# Run extraction
extracted = extract_entities(text)

print(f"--- Extracted Entities ---")
print(f"PERSON: {list(extracted['PERSON'])}")
print(f"ORGANIZATION: {list(extracted['ORGANIZATION'])}")
print(f"LOCATION (GPE): {list(extracted['GPE'])}")
```
